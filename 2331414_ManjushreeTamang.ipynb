```python
import matplotlib.pyplot as plt


image = Image.open("/content/drive/MyDrive/Colab Notebooks/Workshop1/camera_man.jpg").convert("L")

image_array = np.array(image)
print(image_array.shape)

height, width = image_array.shape

data = image_array.copy()

plt.imshow(image_array, cmap='gray')
plt.show()
```

```python
mean = np.mean(data, axis=0)
centered_data = data - mean
```

```python
centered_data.shape
```

```python
cov_matrix = np.cov(centered_data, rowvar=False)
```

```python
cov_matrix.shape
```

```python
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
explained_variance_ratio = eigenvalues/np.sum(eigenvalues)
plt.plot(np.cumsum(explained_variance_ratio))
plt.title("Cumulative Explained Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

```python
k = 40
components = eigenvectors[:, :k]
```

```python
compressed_data = np.dot(centered_data, components)
```

```python
decompressed_data = np.dot(compressed_data, components.T) +mean
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(image_array, cmap="gray")
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(compressed_data, cmap="gray", aspect="auto")
plt.title(f"Compressed Image (k={k})")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(decompressed_data, cmap="gray")
plt.title("Decompressed Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# Load and display grey scale image
Grey_image = Image.open("/content/drive/MyDrive/Colab Notebooks/Workshop1/lenna_image.png").convert("L")
plt.imshow(Grey_image, cmap="gray")
plt.axis("off")
plt.title("Greyscale Image")
plt.show()
```

```python
# Extract and display the middle section of the image (150 pixels)

Grey_image_array = np.array(Grey_image)
# converting grayscale image into array

# get the image dimensions
height, width = Grey_image_array.shape[:2]

# calculate center coordinates
x,y = width // 2, height // 2

# calculate start and end coordinates
start_x, end_x = x - 150//2, x + 150//2
start_y, end_y = y - 150//2, y + 150//2

# extract center 150x150 pixels
center_crop = Grey_image_array[start_y:end_y, start_x:end_x]

# display center top
plt.imshow(center_crop, cmap = "gray")
plt.axis("off")
plt.show()
```

```python
# apply threshold to image

threshold_image = np.where(Grey_image_array < 100, 0, 255).astype(np.uint8)
plt.imshow(threshold_image, cmap = "gray")
plt.axis("off")
plt.show()
```

```python
# rotate the image in 90 degree clockwise direction

rotate_image = Grey_image.rotate(-90, expand = True)
plt.imshow(rotate_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# convert image into RGB image

rgb_array = np.zeros((Grey_image_array.shape[0], Grey_image_array.shape[1], 3), dtype=np.uint8)

# assign grayscle image to all three channels
rgb_array[:, :, 0] = Grey_image_array #red
rgb_array[:, :, 1] = Grey_image_array #green
rgb_array[:, :, 2] = Grey_image_array #blue

plt.imshow(rgb_array)
plt.axis("off")
plt.title("Converted RGB Image")
plt.show()
```

```python
# image compression and decompression using PCA

img = Image.open("/content/drive/MyDrive/Colab Notebooks/Workshop1/camera_man.jpg").convert("L")
img_array = np.array(img)
height, width = img_array.shape
print(height, width)
data = img_array.copy()
plt.imshow(img_array, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# standardize data

mean = np.mean(data, axis=0)
#compute the mean of each column(feature)
center_data = data - mean
center_data.shape
```

```python
# calculate the co-variance matrix of standardize data

cov_matrix = np.cov(center_data, rowvar=False)
cov_matrix.shape
```

```python
# eigen decomposition

eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
# identify principle components
variance_ratio = eigenvalues/np.sum(eigenvalues)
plt.plot(np.cumsum(variance_ratio))
plt.title("Cumulative explained Variance")
plt.xlabel("Number of components")
plt.ylabel("Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

```python
from itertools import compress
# reconstruction and experiment

# create a figure with subplot
k = [10, 20, 30, 40, 50, 80, 100]
plt.figure(figsize=(8,8))
plt.subplot(3,3,1)
plt.imshow(img_array, cmap="gray")
plt.title("Original Image")
plt.axis("off")

for i, k in enumerate(k):
  components = eigenvectors[:, :k]
  compressed_data = np.dot(center_data, components)
  decompressed_data = np.dot(compressed_data, components.T) + mean

  plt.subplot(3,3,i+2)
  plt.imshow(decompressed_data, cmap="gray")
  plt.title(f"{k} Components")
  plt.axis("off")

plt.tight_layout()
plt.show()
```
